```
data_path = "s3://full-stack-bigdata-datasets/Big_Data/Project_Steam/steam_game_output.json"
steam_df = spark.read.json(data_path)
```

```
(steam_ext.selectExpr("publisher as publisher", "name as name")).groupBy('publisher').count().sort('count', ascending=False).show(5)
```

```
# transform
steam_clean = steam_ext \
    .withColumn('linux', F.col('platforms.linux')) \
    .withColumn('mac', F.col('platforms.mac')) \
    .withColumn('windows', F.col('platforms.windows')) \
    .withColumn('rating', F.col('positive') / (F.col('positive') + F.col('negative'))) \
    .withColumn('release_date', F.to_date(F.unix_timestamp(F.col('release_date'), 'yyyy/M/d').cast("timestamp"))) \
    .withColumn('price', F.col('price').cast(DataTypes.FloatType())) \
    .withColumn('initial_price', F.col('initialprice').cast(DataTypes.FloatType())) \
    .withColumn('discount', F.col('discount').cast(DataTypes.FloatType())) \
    


```

```
steam_clean = steam_clean \
    .drop('platforms') \
    .drop('tags') \
    .drop('categories') \
    .drop('website') \
    .drop('header_image')
```

```
display(steam_clean.limit(9))
```

```
steam_clean = (steam_clean
    .withColumn('owners', F.split(F.col('owners'), ' ')) 
    .withColumn('owners', 
        F.expr("cast(regexp_replace(owners[0], ',', '') as int) + cast(regexp_replace(owners[2], ',', '') as int)") / 2)
    .withColumn('revenue', F.col('owners') * F.col('price'))
)
```

```
def explode_column(steam_clean, column_name):
    return steam_clean.withColumn(column_name, F.explode(F.split(F.col(column_name), ',')))

steam_clean = explode_column(steam_clean, 'genre')
steam_clean = explode_column(steam_clean, 'languages')
```

#### Are there years with more releases? Were there more or fewer game releases during the Covid, for example?

```
data_with_year = steam_clean.withColumn('release_year', F.year('release_date'))

# appid by year
appid_per_year = data_with_year.groupBy('release_year').count()
appid_per_year.orderBy('release_year').show()

# COVID-19 (2020-2022)
covid_period = data_with_year.filter((F.col('release_year') >= 2020) & (F.col('release_year') <= 2022))
appid_covid = covid_period.groupBy('release_year').count().orderBy('release_year')

appid_covid.show()
```

#### How are the prizes distributed? Are there many games with a discount?

```
discounted_games = steam_clean.filter(F.col('price') < F.col('initial_price'))

total_games = steam_clean.count()

num_discounted_games = discounted_games.count()

percentage_discounted = (num_discounted_games / total_games) * 100 if total_games > 0 else 0

print(f"Total number of games: {total_games}")
print(f"Number of games with a discount: {num_discounted_games}")
print(f"Percentage of games with a discount: {percentage_discounted:.2f}%")
```

```
df = steam_clean.selectExpr(
    "appid as appid", 
    "price", 
    "(initial_price - price) as discount" 
).orderBy(F.desc("discount")) 

display(df.limit(9))
```

Are there many games prohibited for children under 16/18?

```

total_unique_games = steam_clean.select("appid").distinct().count()

prohibited_under_16 = steam_clean.filter(F.col('required_age')> 16 ).select("appid").distinct().count()
prohibited_under_18 = steam_clean.filter(F.col('required_age') > 18 ).select("appid").distinct().count()

percentage_under_16 = (prohibited_under_16 / total_unique_games * 100) if total_unique_games > 0 else 0
percentage_under_18 = (prohibited_under_18 / total_unique_games * 100) if total_unique_games > 0 else 0

print(f"Percentage of unique games prohibited for children under 16: {percentage_under_16:.2f}%")
print(f"Percentage of unique games prohibited for children under 18: {percentage_under_18:.2f}%")
```

#### Are there any genres that have a better positive/negative review ratio?

```
genre_stats = steam_clean.groupBy('genre').agg(
    F.mean('rating').alias('average_rating'),
    F.count('rating').alias('games_in_genre')
)
top_genres = genre_stats.orderBy(F.desc('average_rating')).limit(10)
least_genres = genre_stats.orderBy('average_rating').limit(10)

display(top_genres, least_genres)
```

```
display(
    steam_clean.groupBy('genre').agg(F.mean('rating').alias('average_rating'), F.count('rating').alias('games_in_genre')).orderBy('average_rating', ascending=False).limit(10),
    steam_clean.groupBy('genre').agg(F.mean('rating').alias('average_rating'), F.count('rating').alias('games_in_genre')).orderBy('average_rating', ascending=True).limit(10)
)
```

#### What are the most lucrative genres? üí∞üí∏

```

genre_revenue_stats = steam_clean.groupBy('genre').agg(
    F.mean('revenue').alias('average_revenue')
)
top_revenue_genres = genre_revenue_stats.orderBy(F.desc('average_revenue')).limit(10)
#least_revenue_genres = genre_revenue_stats.orderBy('average_revenue').limit(10)

display(top_revenue_genres)
```

```
display(steam_clean.groupBy('genre').mean('revenue').orderBy('avg(revenue)', ascending=False).limit(10),
        steam_clean.groupBy('genre').mean('revenue').orderBy('avg(revenue)', ascending=True).limit(10))
```

## Platform analysis üíøüéÆüì∫

#### Are most games available on Windows/Mac/Linux instead?

```
total_games = steam_clean.count()
platforms = ['windows', 'linux', 'mac']

for platform in platforms:
    count_platform = steam_clean.filter(F.col(platform) == True).count()
    percentage = (count_platform / total_games * 100) if total_games > 0 else 0
    print(f"{platform}: {percentage:.2f} %")
```
